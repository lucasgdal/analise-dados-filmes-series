# Selecionando tipos de entreterimento

```python
import pandas as pd 
```

```python
dados = pd.read_csv('base_limpa.csv', sep=';')
```

```python
dados.head(5)
```

```python
selecao_filme = dados['type'] == 'Movie'
selecao_series = dados['type'] == 'TV Show'
```

```python
dados[selecao_filme].shape[0]
```

```python
dados[selecao_series].shape[0]
```

# Selecionando somente os filmes

```python
filmes = dados[selecao_filme]
```

```python
filmes.index = range(filmes.shape[0])
```

```python
filmes
```

```python
filmes.to_csv('filmes.csv', index=False)
```

# Selecionando somente series

```python
series = dados[selecao_series]
```

```python
series.index = range(series.shape[0])
```

```python
series.to_csv('series.csv',index=False)
```
