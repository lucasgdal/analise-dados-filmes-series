## Tratamento dos dados 


```python
import pandas as pd 
```

```python
base = pd.read_csv('netflix.csv')
```

```python
base.head(5)
```

```python
base.info()
```

```python
base.isnull().sum()
```

```python
# porcentagem dos dados nulos no dataset

base.isnull().sum()/base.shape[0]*100
```

```python
# como a porcentagem de dados na coluna 'director' é muito alta achei melhor colocar os dados faltante como desconhecido

base.director = base.director.fillna('Unknown',inplace=False)

base.isnull().sum()
```

```python
base.isnull().sum()/base.shape[0]*100
```

```python
base_tratada = base.dropna(inplace=False, axis=0)
```

```python
base_tratada.head(10)
```

```python
base_tratada.to_csv('base_limpa.csv',sep=';',index=False)
```
